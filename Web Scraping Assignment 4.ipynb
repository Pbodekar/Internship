```python
# Creating Lists
details = []
Name = []
Designation = []
Company = []
Skill = []
Location = []

# Extracting urls for recruiter
recr_url = driver.find_elements_by_xpath('//a[@class="ellipsis"]')

for i in recr_url:
    details.append(i.get_attribute("href"))
    
# Extracting data from urls
for recr_url in details:
    driver.get(recr_url)
    time.sleep(2)
    
# Extracting Name from url   
    try:
        name = driver.find_element_by_xpath('//span[@class="fl ellipsis"]')
        Name.append(name.text)
    except NoSuchElementException:
        Name.append('-')
    
    
    time.sleep(2)
# Extracting Designation from url
    try:
        designation = driver.find_element_by_xpath('//span[@class="ellipsis clr"]')
        Designation.append(designation.text)
    except NoSuchElementException:
        Designation.append('-')
        
        
    time.sleep(2)
# Extracting Company Name from url
    try:
        company = driver.find_element_by_xpath('//i[@class="cmpIcn"]')
        Company.append(company.text)
    except NoSuchElementException:
        Company.append('-')
        
        
    time.sleep(2)
    
# Extracting Skill for hiring from url   
    try:
        skill = driver.find_element_by_xpath('//div[@class="hireSec highlightable"]')
        Skill.append(skill.text)
    except NoSuchElementException:
        Skill.append('-')
        
        
    time.sleep(2)
#Extracting Location from url
    try:
        location = driver.find_element_by_xpath('//small[@class="ellipsis"]')
        Location.append(location.text)
    except NoSuchElementException:
        Location.append('-')
        
        
time.sleep(3)
# Make DataFrame
Jobs = pd.DataFrame({'Name':Name,'Designation':Designation,'Company':Company,'Skills they hire for':Skill,'Location':Location})
Jobs
```

```python
# Extracting all data
num = driver.find_elements_by_xpath('/html[1]/body[1]/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/table[1]/tbody[1]/tr/td')
len(num)
```

```python
# Creating List
detail = []
rank = []
book = []
author = []
sales= []
publisher = []
genre = []

# Extracting data from xpath for Rank, Book Name, Author Name, 
for i in num:
    detail.append(i.text)
for i in range(0,len(detail)-1,6):
    rank.append(detail[i])              # Extracting Rank form range function
    book.append(detail[i+1])            # Extracting Book Name form range function
    author.append(detail[i+2])          # Extracting Author Name form range function
    sales.append(detail[i+3])           # Extracting Value Sold form range function
    publisher.append(detail[i+4])       # Extracting Publisher form range function
    genre.append(detail[i+5])           # Extracting Genre form range function

# Make DataFrame
Best_Selling = pd.DataFrame()    
Best_Selling['Rank'] = rank
Best_Selling['Book Name'] = book
Best_Selling['Author Name'] = author
Best_Selling['Valume Sold'] = sales
Best_Selling['Publisher'] = publisher
Best_Selling['Genre'] = genre
Best_Selling
```

Q9 Scrape the details most watched tv series of all time from imdb.com.

```python
# Activating the chrome browser
driver = webdriver.Chrome(r'C:\Users\praja\Downloads\app\chromedriver_win32\chromedriver.exe')
```

```python
# Extracting Url
url = 'https://www.imdb.com/list/ls095964455/'
driver.get(url)
driver.maximize_window()        # Maximizing window
```

```python
# Creating Lists
Name = []
Year = []
Genre = []
Run_time = []
Rating = []
Votes = []

# Extracting all data 
name =  driver.find_elements_by_xpath('//h3[@class="lister-item-header"]')

# Extracting Name
for i in name:
    Name.append(i.text.split('.')[1])

# Extracting Year
year = driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')

for i in year:
    Year.append(i.text)

# Extracting Genre
genre = driver.find_elements_by_xpath('//span[@class="genre"]')
for i in genre:
    Genre.append(i.text)

# Extracting Run
run = driver.find_elements_by_xpath('//span[@class="runtime"]')
for i in run:
    Run_time.append(i.text)

# Extracting Rating
rate = driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]')
for i in rate:
    Rating.append(i.text)
    
# Extracting Voting
vote = driver.find_elements_by_xpath('//span[@name="nv"]')
for i in vote:
    Votes.append(i.text)

time.sleep(2)

# Checking len for all attribute
len(Name),len(Year),len(Genre),len(Run_time),len(Rating),len(Votes)

# Make DataFrame
Series = pd.DataFrame({'Name':Name,'Year Span':Year,'Genre':Genre,'Run Time':Run_time,'Rating':Rating,'Votes':Votes})
Series
```
